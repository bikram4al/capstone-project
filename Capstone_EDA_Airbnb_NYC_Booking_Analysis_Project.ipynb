## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# creating a copy of current dataset and assigning to df
df=dataset.copy()
# Handle missing values
dataset['reviews_per_month'].fillna(0, inplace=True)
dataset.dropna(subset=['last_review'], inplace=True)

# Verify missing values are handled
missing_values = dataset.isnull().sum()
print("Missing Values after handling:\n", missing_values)
# finding out new shape of dataset after handling missing value
print("New shape of dataset after handling missing value:\n",df.shape)
````

````python
# Ensure consistency in 'neighbourhood_group' and 'room_type' values
df['neighbourhood_group'] = df['neighbourhood_group'].str.strip()
df['room_type'] = df['room_type'].str.strip()

# Verify unique values in 'neighbourhood_group' and 'room_type'
print("Unique Neighbourhood Groups:", df['neighbourhood_group'].unique())
print("Unique Room Types:", df['room_type'].unique())
````

````python
from datetime import datetime

# Convert 'last_review' to datetime
df['last_review'] = pd.to_datetime(df['last_review'])

# Create new features: days since last review and listing age
df['days_since_last_review'] = (datetime.now() - df['last_review']).dt.days
df['listing_age'] = datetime.now().year - df['last_review'].dt.year

# Create a binary feature: 'is_high_availability'
df['is_high_availability'] = df['availability_365'].apply(lambda x: 1 if x > 200 else 0)

# Create a binary feature: 'is_expensive_listing' based on price
df['is_expensive_listing'] = df['price'].apply(lambda x: 1 if x > df['price'].median() else 0)

# Create price categories
price_bins = [0, 100, 200, 300, 500, 10000]
price_labels = ['Budget', 'Affordable', 'Midrange', 'Premium', 'Luxury']
df['price_category'] = pd.cut(df['price'], bins=price_bins, labels=price_labels)

# Create review count categories
review_bins = [0, 10, 50, 100, 500, 10000]
review_labels = ['Few', 'Moderate', 'Good', 'High', 'Extensive']
df['review_category'] = pd.cut(df['number_of_reviews'], bins=review_bins, labels=review_labels)

# Verify new features
print(df[['days_since_last_review', 'listing_age', 'is_high_availability', 'is_expensive_listing', 'price_category', 'review_category']].head())
````

````python
#comparing availability across different neighbourhoods
neighbourhood_agg_availability = neighbourhood_agg.sort_values(by='mean_availability', ascending=False).head(10)
print(" Demand by neighbourhood :\n",neighbourhood_agg_availability)
````

````python
# Aggregate by neighbourhood group
neighbourhood_group_agg = df.groupby('neighbourhood_group').agg({
    'price': ['mean', 'median', 'std'],
    'number_of_reviews': ['mean', 'median'],
    'availability_365': ['mean', 'sum'],
    'calculated_host_listings_count': ['mean', 'sum'],
    'id': 'count'
}).reset_index()

# Rename columns for readability
neighbourhood_group_agg.columns = ['neighbourhood_group', 'mean_price', 'median_price', 'std_price',
                             'mean_reviews', 'median_reviews', 'mean_availability', 'total_availability',
                             'mean_host_listings', 'total_host_listings', 'total_listings']

print(neighbourhood_group_agg.head())
````

````python
# Aggregate by room type
room_type_agg = df.groupby('room_type').agg({
    'price': ['mean', 'median'],
    'number_of_reviews': ['mean', 'median'],
    'availability_365': ['mean', 'sum'],
    'id': 'count'
}).reset_index()

# Rename columns for readability
room_type_agg.columns = ['room_type', 'mean_price', 'median_price',
                         'mean_reviews', 'median_reviews', 'mean_availability', 'total_availability',
                         'total_listings']

print(room_type_agg)
````

````python
# Aggregate by host ID
host_agg = df.groupby('host_id').agg({
    'price': ['mean', 'median'],
    'number_of_reviews': ['mean', 'median'],
    'availability_365': ['mean', 'sum'],
    'id': 'count'
}).reset_index()

# Rename columns for readability
host_agg.columns = ['host_id', 'mean_price', 'median_price',
                    'mean_reviews', 'median_reviews', 'mean_availability', 'total_availability',
                    'total_listings']

# Calculate the percentage of hosts with multiple listings
multi_listing_hosts = len(host_agg[host_agg['total_listings'] > 1]) / len(host_agg) * 100
print("Percentage of Hosts with Multiple Listings:", multi_listing_hosts)
print(host_agg.head())
````

````python
# Aggregate based on last review year
df['last_review_year'] = df['last_review'].dt.year

review_year_agg = df.groupby('last_review_year').agg({
    'price': ['mean', 'median'],
    'number_of_reviews': ['mean', 'median'],
    'availability_365': ['mean', 'sum'],
    'id': 'count'
}).reset_index()

# Rename columns for readability
review_year_agg.columns = ['last_review_year', 'mean_price', 'median_price',
                           'mean_reviews', 'median_reviews', 'mean_availability', 'total_availability',
                           'total_listings']

print(review_year_agg)
review_year_agg.shape
````

````python
# Detect outliers in 'price' using Z-score
from scipy.stats import zscore

df['price_zscore'] = zscore(df['price'])

# Identify outliers
price_outliers_zscore = df[df['price_zscore'].abs() > 3]
print("Number of Price Outliers:", len(price_outliers_zscore))

# Remove outliers for further analysis
data_cleaned = df[df['price_zscore'].abs() <= 3]
print("Shape of Data after Removing Price Outliers:", data_cleaned.shape)
````

````python
# cross feature analysis
# price vs availability
price_availability_corr = df['price'].corr(df['availability_365'])
print("Correlation between Price and Availability:", price_availability_corr)
````
